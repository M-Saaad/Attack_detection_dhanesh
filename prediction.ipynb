```python
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_excel('./PublicPrivateSectors.xlsx')
df.head()
```

```python
df = df.drop(['Category', 'Author', 'DateTime', 'Href', 'Headline', 'Short_Summary', 'Long_Summary', 'Organization'], axis=1)
df.head()
```

```python
print(df['Locations'].value_counts())
```

```python
df = df[df['Locations'] != ' ']
df = df[df['Organization_Type'] != ' ']
df = df[df['PublicPrivate'] != ' ']
df
```

```python
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

df['Locations']= label_encoder.fit_transform(df['Locations'])
df['Locations'].unique()
print("Locations = ", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))

df['Organization_Type']= label_encoder.fit_transform(df['Organization_Type'])
df['Organization_Type'].unique()
print("Organization_Type = ", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))

df['Attacks']= label_encoder.fit_transform(df['Attacks'])
df['Attacks'].unique()
print("Attacks = ", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))

df['PublicPrivate']= label_encoder.fit_transform(df['PublicPrivate'])
df['PublicPrivate'].unique()
print("PublicPrivate = ", dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))

df
```

```python
X = df.iloc[:,[0,1,3]]
y = df.iloc[:,[2]]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train, y_train)

y_pred = reg.predict(X_train)

# Training Score
print("Training Accuracy:", reg.score(X_train, y_train)*100)
```

```python
from sklearn.linear_model import SGDRegressor

reg = SGDRegressor()
reg.fit(X_train, y_train)

y_pred = reg.predict(X_train)

# Training Score
print("Training Accuracy:", reg.score(X_train, y_train)*100)
```

```python
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression(multi_class='multinomial', solver='lbfgs')

reg.fit(X_train,y_train)

y_pred = reg.predict(X_train)

# Training Score
print("Training Accuracy:", reg.score(X_train, y_train)*100)
```

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor()

reg.fit(X_train,y_train)

y_pred = reg.predict(X_train)

# Training Score
print("Training Accuracy:", reg.score(X_train, y_train)*100)
```
